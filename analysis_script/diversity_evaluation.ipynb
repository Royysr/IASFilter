```python
from rdkit import Chem
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import pandas as pd
import os
from rdkit.Chem import MolStandardize,rdmolops,MACCSkeys,ChemicalFeatures,AllChem,Descriptors,Lipinski
path = os.getcwd()
database = pd.read_csv(f'{path}/datasets/example_train.csv').iloc[:,:2]         # please check molecular data file. It should be in SMILES-Labels formatting, which contains two columns.
smiles = list(database.iloc[:, 0])
labels = database.iloc[:,1]
plt.figure(figsize=(8, 6), dpi=300)
plt.hist(labels,
         bins=30,
         color='#AADCA9',
         edgecolor='black')
plt.xticks(fontsize=28, fontname='Arial', fontweight='bold')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{int(x)}'))
plt.yticks(fontsize=28, fontname='Arial', fontweight='bold')
plt.title(f'Labels Distribution',fontsize=29, fontname='Arial', fontweight='bold')         ## fontsize=33  29
plt.tight_layout()
plt.show()
```

```python
def cal_rings(smi_list):
    ring_list = []
    for smi in smi_list:
        m = Chem.MolFromSmiles(smi)
        ring_list.append(m.GetRingInfo().NumRings())
    print('maximum', max(ring_list),
          'minimum', min(ring_list),
          'average', sum(ring_list) / len(ring_list))
    return ring_list
ring = cal_rings(smiles)
plt.figure(figsize=(8, 6), dpi=300)
bins = range(min(ring), max(ring) + 2)
plt.hist(ring,
         bins=bins,
         color='#206FB6',
         edgecolor='black',
         align='left',
         rwidth=0.8)
plt.gca().yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, pos: f'{int(x)}')
)
step = 1
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(step))
plt.gca().xaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, pos: f'{int(x)}' if x % step == 0 else '')
)
plt.xticks(fontsize=28, fontname='Arial', fontweight='bold')
plt.yticks(fontsize=28, fontname='Arial', fontweight='bold')
plt.title(f'Number of Rings',
          fontsize=29, fontname='Arial', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
def cal_atom_num(smi_list):
    count_list = []
    for smi in smi_list:
        m = Chem.AddHs(Chem.MolFromSmiles(smi))
        c = m.GetNumAtoms()
        count_list.append(c)
    print('maximum', max(count_list),
          'minimum', min(count_list),
          'average', sum(count_list) / len(count_list))
    return count_list
count = cal_atom_num(smiles)
plt.figure(figsize=(8, 6), dpi=500)
plt.hist(count,
         bins=20,
         color='#6BADD7',
         edgecolor='black',
         align='mid')

plt.xticks(fontsize=30, fontname='Arial', fontweight='bold')
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(nbins=6, integer=True))
plt.gca().yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, pos: f'{int(x)}')
)
plt.xticks(fontsize=28, fontname='Arial', fontweight='bold')
plt.yticks(fontsize=28, fontname='Arial', fontweight='bold')

plt.title(f'Number of Atoms',
          fontsize=29, fontname='Arial', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# the number of rotatable bonds
def get_rotatable_bonds(smi_list):
    rotatable_bonds_num = []
    for smi in smi_list:
        mol = Chem.AddHs(Chem.MolFromSmiles(smi))
        rotatable_bonds_num.append(Lipinski.NumRotatableBonds(mol))
    print('maximum', max(rotatable_bonds_num),
          'minimum', min(rotatable_bonds_num),
          'average', sum(rotatable_bonds_num) / len(rotatable_bonds_num))
    return rotatable_bonds_num
count_bonds = get_rotatable_bonds(smiles)
plt.figure(figsize=(8, 6), dpi=500)
bins = 10
plt.hist(count_bonds,
         bins=bins,
         color='#FC9171',
         edgecolor='black',
         align='mid')
plt.xticks(fontsize=30, fontname='Arial', fontweight='bold')
plt.gca().yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, pos: f'{int(x)}')
)
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.title(f'Rotatable Bonds',
          fontsize=29, fontname='Arial', fontweight='bold')
plt.xticks(fontsize=28, fontname='Arial', fontweight='bold')
plt.yticks(fontsize=28, fontname='Arial', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
def get_element_distribution_per_molecule(smi_list):
    elements_per_mol = []
    for smi in smi_list:
        mol = Chem.MolFromSmiles(smi)
        mol_elements = set([atom.GetSymbol() for atom in mol.GetAtoms()])
        elements_per_mol.extend(mol_elements)
    return Counter(elements_per_mol)
element_counts = get_element_distribution_per_molecule(smiles)
elements, counts = zip(*element_counts.items())
plt.figure(figsize=(8, 6), dpi=500)
plt.bar(elements,
        counts,
        color='#FDDFD0',
        edgecolor='black')
plt.xticks(fontsize=25, fontname='Arial', fontweight='bold',
           rotation=45, ha='right')
plt.gca().yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, pos: f'{int(x)}')
)
plt.yticks(fontsize=28, fontname='Arial', fontweight='bold')
plt.title(f"Element Distribution",
          fontsize=29, fontname='Arial', fontweight='bold')
plt.tight_layout()
plt.show()
```
